```python
# shopping list app
# step 1: initilize an empty shopping list
shopping_list = []

#step 2: define the main menu
def show_menu():
    print("\n---- Shopping List Menu ----")
    print("1. View list")
    print("2. Add item to list")
    print("3. Remove item from list")
    print("4. edit the list")
    print("5. save the list")
    print("6. clear List")
    print("7. Quit")

#step 3: main program loop
while True:
    show_menu()
    choice = input("Enter your choice (1-7): ")

    if choice == "1":
        if not shopping_list:
            print("Your shopping list is empty.")
        else:
            for index, item in enumerate(shopping_list):
                print(f"{index+1}. {item}")
    elif choice == "2":
        item = input("Enter the item you want to add: ")
        shopping_list.append(item)
        print(f"{item} has been added to the list.")

    elif choice == "3":
        if not shopping_list:
            print("Your shopping list is empty.")

        else:
            item = input("Enter the item you want to remove: ")
            if item in shopping_list:
                shopping_list.remove(item)
                print(f"{item} has been removed from the list.")

            else:
                print(f"{item} is not in the list.")

    elif choice == "4":
        if not shopping_list:
            print("Your shopping list is empty.")

        else:
            item = input("Enter the item you want to edit: ")
            if item in shopping_list:
                index = shopping_list.index(item)
                new_item = input("Enter the new item: ")
                shopping_list[index] = new_item
                print(f"{item} has been replaced with {new_item} in the list.")

            else:
                print(f"{item} is not in the list.")


    elif choice == "5":
        filename = input("Enter the filename to save the list: ")
        with open(filename, "w") as file:
            for item in shopping_list:
                file.write(item + "\n")
        print(f"The list has been saved to {filename}.")

    elif choice == "6":
        shopping_list.clear()
        print("Your shopping list has been cleared.")

    elif choice == "7":
        print("Goodbye!")
        break

    else:
        print("Invalid choice. Please enter a number between 1 and 5.")

```
